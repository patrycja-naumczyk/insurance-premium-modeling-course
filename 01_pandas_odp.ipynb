# <a id='method_chaining'>Method chaining </a>
Stwórz dataframe df1 jako podzbiór df w następujących krokach: 
* weź dataframe df
* wybierz wszystkie kursy do/z Gdańska 
* usuń kolumny *zwiekszenie_opoznienia*, *zwiekszenie_opoznienia_assign*, *czy_start_stolica*, *podtyp_pendolino* 
* stwórz kolumnę *czy_start_gdansk*
* stwórz kolumnę *srednie_opoznienie_typ* przypisujące średnie opóźnienie na mecie dla danego typu pociągu

```python
# Zadanie 1
cols_to_drop = ['zwiekszenie_opoznienia', 'zwiekszenie_opoznienia_assign', 'czy_start_stolica', 'podtyp_pendolino']

df1 = (
    df
    .loc[lambda df_: ( df_['Stacja początkowa:'].str.contains("Gdańsk")) | (df_['Stacja końcowa:'].str.contains("Gdańsk"))]
    .drop(columns=cols_to_drop)
    .assign(
        czy_start_gdansk = lambda df_: np.where(
            df_['Stacja początkowa:'].str.contains("Gdańsk") ,
            1,
            0
        )
    )
    .pipe(lambda df_: pd.merge(
        df_,
        (
            df
            .loc[
                ( df['Stacja początkowa:'].str.contains("Gdańsk")) | (df['Stacja końcowa:'].str.contains("Gdańsk")), 
                ['Liczba minut opóźnienia na mecie:', 'Rodzaj pociągu:']
            ]
            .groupby(by='Rodzaj pociągu:') 
            .mean()
            .reset_index()
            .rename(columns={
                'Liczba minut opóźnienia na mecie:' : 'srednie_opoznienie_typ'
            })
        ),
        on='Rodzaj pociągu:',
        how='left'
    ))
)
df1
```
