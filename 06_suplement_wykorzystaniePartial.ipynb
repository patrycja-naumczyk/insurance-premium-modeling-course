```python
from functools import partial
import numpy as np
```

```python
# własna funkcja wyliczająca odległość euklidesowską
def euklides(a, b):
    x1, y1 = a
    x2, y2 = b
    return np.sqrt((x1-x2)**2 + (y1-y2)**2)
```

```python
# definicja punktu odniesienia
target = (0,0)

# definicja nowej funkcji ze stałym parametrem a
euk_target = partial(euklides, a=target)
```

```python
# stworzenie przykładowej listy punktów do policzenia
points_list = [(x, y) for x in range(-5, 6) for y in range(-5, 6)]
print(
    f"Dla targetu: {target}:\n"    
)

# wyświetlenie odległości euklidesowskich
for point in points_list:
    print(
        f"Punkt: {point}, odległość euklidesowska od targetu: {euk_target(b=point):.3}"
    )
```
